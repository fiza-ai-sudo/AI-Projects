```python
# Smart_chatbot
# import libraries
memory={}
!pip install transformers
# Rule based replies
def rule_based_reply(user_input):
    user_input=user_input.lower()
# store name
    if "my name is" in user_input:
        name=user_input.replace("my name is","").strip()
        memory["name"]=name
        return f"Nice to meet you {name}."
# call your name
    elif "what is my name" in user_input:
        if "name" in memory:
            return f"your name is {memory['name']}."
        else:
            return "i don't know your name yet"
    if "hi" in user_input or "hello" in user_input:
        return "Hello! 🤗 how are you?"
    elif "i am good thank you 🤗 what about you??" in user_input:
        return "I am fine too."
    elif "i am bored" in user_input or "bored" in user_input:
        return "Oh 🥺 let's talk say something what you want"
    elif "what you can do feel me refresh?" in user_input:
        return "we can play games"
    elif "Goodbye 👋" in user_input or "Allah Hafiz 🥰" in user_input:
        return "Goodbye 👋 Fiza"
    else:
        return None
# ML based replies
from transformers import pipeline
nlp = pipeline("text-generation", model="distilgpt2")

def ml_reply(user_input):
    response = nlp(user_input,max_length=50, num_return_sequences=1)
    return response[0]["generated_text"]
# Chatbot run
while True:
    user=input("You: ")
    rule=rule_based_reply(user)
    if rule:
        print("Bot: ",rule)
    else:
        print("Bot: ",ml_reply(user))
```
